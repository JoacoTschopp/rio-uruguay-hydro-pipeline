# Carga Historico Temp Aeropuerto

```python
bronze_schema = spark.table("weather.bronze.metar").schema

historical_df = (
    spark.read
         .schema(bronze_schema)
         .option("multiLine", True)
         .json("/Volumes/weather/raw/noaa_volume/json/historical/")
)

historical_df.write \
    .format("delta") \
    .mode("append") \
    .saveAsTable("weather.bronze.metar")
```
