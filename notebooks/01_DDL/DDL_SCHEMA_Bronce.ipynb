# Schemas

```python
spark.sql("CREATE CATALOG IF NOT EXISTS weather")

spark.sql("USE CATALOG weather")

spark.sql("CREATE SCHEMA IF NOT EXISTS raw")
spark.sql("CREATE SCHEMA IF NOT EXISTS bronze")
spark.sql("CREATE SCHEMA IF NOT EXISTS silver")
spark.sql("CREATE SCHEMA IF NOT EXISTS gold")
```

# Tabla Bronze para Temperatura en aeropuertos

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS weather.bronze.metar
USING DELTA
AS
SELECT *
FROM json.`/Volumes/weather/raw/noaa_volume/json/daily/`
LIMIT 0
""")
```

# Tabla Bronze para Niveles de ANA (Target futuro!!)

```python
# Load a sample record to infer schema
sample_df = spark.read.option("multiLine", True).json("/Volumes/weather/raw/ana_volume/json/serie_74100000.json")
sample_schema = sample_df.schema

# Create empty table with inferred schema
spark.createDataFrame([], sample_schema).write.format("delta").saveAsTable("weather.bronze.nivel_ana")
```

```python
%sql
describe weather.bronze.nivel_ana
```
